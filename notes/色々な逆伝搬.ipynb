### 1. 平均の微分
以下の式を$x_i$について微分する

$$
μ = \frac{1}{N}\sum_{i=1}^\N x_i
$$

#### 式の展開

$$
μ = \frac{1}{N}x_1 + \frac{1}{N}x_2 + \frac{1}{N}x_3 ... + \frac{1}{N}x_N
$$

#### $x_i$について微分
$$
\begin{align*}
\frac{\partial μ}{\partial x_i}
&=\frac{1}{N} \cdot 1 + \frac{1}{N} \cdot 0 + \frac{1}{N} \cdot 0 ... \frac{1}{N} \cdot 0 \\
&= \frac{1}{N}
\end{align*}
$$

### 2. $\sum_{i=1}^\N x_i - μ$の微分

#### 式の展開

$$
A = (x_1 - μ) + (x_2 - μ) + (x_3 - μ) ... + (x_N - μ)
$$


#### $x_i$について微分
$$
\begin{align*}
\frac{\partial A}{x_i}
&=(1 - \frac{1}{N}) - \frac{1}{N} - \frac{1}{N} ... -\frac{1}{N} \\
&= (1 - \frac{1}{N}) - (1 - \frac{1}{N}) \\
&= 0
\end{align*}
$$

##### 右辺の計算について$- \frac{1}{N} - \frac{1}{N} ... \frac{1}{N}$
$$
\begin{align*}
1 - \frac{1}{N} 
&= - \frac{1}{N} - \frac{1}{N} ... -\frac{1}{N} \\
&= (N - 1) \cdot -\frac{1}{N} \\
&= 1 - \frac{1}{N}
\end{align*}
$$
